```python
import matplotlib.pyplot as plt
import os
import glob
import cv2
import seaborn as sns
import pandas as pd
```

```python
path = "dataset/images"
train_img_paths = glob.glob(os.path.join(path, "train", "*.jpg"))
val_img_paths = glob.glob(os.path.join(path, "val", "*.jpg"))
test_img_paths = glob.glob(os.path.join(path, "public_test", "*.jpg"))

train_label_paths = [k.replace("images", "labels").replace("jpg", "txt") for k in train_img_paths]
val_img_paths = [k.replace("images", "labels").replace("jpg", "txt") for k in val_img_paths]
test_img_paths = [k.replace("images", "labels").replace("jpg", "txt") for k in test_img_paths]
```

```python
class_names = ['no_mask', 'mask', 'incorrect_mask']
img_paths = [train_img_paths, val_img_paths, test_img_paths]
label_paths = [train_label_paths, val_img_paths, test_img_paths]

dataset_type = ["train", "val", "test"]
```

```python
for k, v in zip(dataset_type, img_paths):
    print(f"{k} dataset have {len(v)} images")
```

```python
# check img size
img_shape = None
for k in img_paths:
    for img_path in k:
        img = cv2.imread(img_path)
        cur_img_shape = img.shape[:2] # h, w
        print(f"{img_path} -> {cur_img_shape}")
```

```python
def read_label_file(file):
    rows = []
    with open(file, "r") as f:
        for line in f:
            line = line.rstrip().split(" ") # class name, x, y, w, h
            line = list(map(float, line))
            rows.append(line)
    return rows

def get_size_from_label_path(label_file):
    img_file = label_file.replace("labels", "images").replace("txt", "jpg")
    img = cv2.imread(img_file)
    img_size = img.shape[:2] # h, w
    return img_size
    
def scale_yolo_coord(bbox, img_size):
    xc, yc, w, h = bbox
    img_h, img_w = img_size
    xc_s, yc_s, w_s, h_s = xc * img_w, yc * img_h, w * img_w, h * img_h
    xt_s, yt_s = xc_s - w_s/2, yc_s - h_s/2
    return [xt_s, yt_s, w_s, h_s]
```

```python
cols = ["img_name", "type", "img_size", "bbox", "class"]
df = {k:[] for k in cols}
```

```python
for label_group, type in zip(label_paths, dataset_type):
    for label_file in label_group:
#         print(label_file)
        objs = read_label_file(label_file)
        
        img_size = get_size_from_label_path(label_file)
        for obj in objs:
            class_name = class_names[int(obj[0])]
            bbox_coord = scale_yolo_coord(obj[1:], img_size)
            row_val = [label_file, type, img_size, bbox_coord, class_name]

            for r, c in zip(row_val, cols):
                df[c].append(r)
```

```python
df = pd.DataFrame(df)
df.head()
```

#### visualise img

```python
def vis_img(img_path, bboxes, classes):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    for box, cls in zip(bboxes, classes):        
        x, y, w, h = list(map(int, box))
        cv2.rectangle(img, (x, y), (x + w, y + h), (36,255,12), 2)
        cv2.putText(img, cls, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)
    
    plt.figure(figsize=(15, 7))
    plt.imshow(img)
    plt.show()
```

```python
df_gb = df.groupby("img_name")
bboxes = df_gb["bbox"].apply(list).to_dict()
classes = df_gb["class"].apply(list).to_dict()
```

```python
img_paths = df_gb["img_name"].groups.keys()

i = 0
for img_path in img_paths:
    boxes = bboxes[img_path]
    cls = classes[img_path]
    img_path = img_path.replace("labels", "images").replace("txt", "jpg")
    vis_img(img_path, boxes, cls)
    i += 1
    if i == 5:
        break
```
